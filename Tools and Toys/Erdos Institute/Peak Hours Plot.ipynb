```python
# Making simple plots of peak hours of the selected bars and cafes
import pandas as pd 
import matplotlib.pyplot as plt

peak_hours_data = pd.read_csv('peakhours-calcsheet.csv')
peak_hours_data.drop(columns=peak_hours_data.columns[0], axis=1,  inplace=True)

cafe_peaks = peak_hours_data.loc[[15]]
bar_peaks = peak_hours_data.loc[[39]]

day_hours = list(range(24))

plt.plot(day_hours, cafe_peaks.T, 'red')
plt.plot(day_hours, bar_peaks.T, 'blue')
plt.xlabel('Time of Day (Hour, ET)')
plt.ylabel('Proportion of Nearby Stores at Peak Traffic')
plt.xlim(0, 23)
plt.legend(['Cafe', 'Bar'])
plt.title('Comparison of Cafes vs Bars: When Are Their Peak Hours?')
plt.show()

# Heat maps of daily active hours - cafes vs bars


# Scatter plot of peak hours - cafes in red, bars in blue
```
